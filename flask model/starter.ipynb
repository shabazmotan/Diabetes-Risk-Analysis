```python
# importing the dataset
import pandas
import numpy
from sklearn import preprocessing
 
df = pandas.read_csv('diabetes.csv')
df.head()
```

```python
col_names = df.columns
 
for c in col_names:
    df = df.replace("?", numpy.NaN)
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
df['BMI']
```

```python
df.dtypes
```

```python
df.loc[df.BMI < 26, 'BMI'] = 0
df.loc[df.BMI >= 26, 'BMI'] = 1
    
```

```python
df.drop(['GenHlth', 'MentHlth', 'PhysHlth', 'Age', 'Education', 'Income'], axis=1, inplace=True)
```
